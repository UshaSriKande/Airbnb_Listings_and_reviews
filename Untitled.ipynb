```python
#paris_listings_neighbourhood = (df.groupby("neighbourhood")["price"].mean())

#paris_listings_neighbourhood = (df.groupby("neighbourhood")["price"].sum())


#paris_listings_neighbourhood = (df.groupby("neighbourhood")["price"].agg(['mean','sum']))


paris_listings_neighbourhood = (df.groupby("neighbourhood").agg({"price":"mean"}))
paris_listings_neighbourhood.head()
```

```python
paris_listings_neighbourhood.sort_values(by="price")
paris_listings_neighbourhood.sort_values(by="price",ascending=[True]).head()

```

```python
paris_listings_accommodates = (df.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price":"mean"}))
paris_listings_accommodates.sort_values("price",inplace=True)
```
